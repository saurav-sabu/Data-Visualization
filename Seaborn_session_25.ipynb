### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# scatter plot -> axes level function
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
# scatter using relplot -> size and hue -> Figure level function
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter")
```

```python
# style semantics
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",hue="sex")
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex",style="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",hue="sex",style="time")
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex",size="size")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",hue="sex",size="size")
```

```python
gap = px.data.gapminder()
gap.head()
```

```python
# line plot
temp_df = gap[gap["country"] == "India"]
sns.lineplot(x="year",y="lifeExp",data=temp_df)
```

```python
# using relpplot
temp_df = gap[gap["country"] == "India"]
sns.relplot(x="year",y="lifeExp",data=temp_df,kind="line")
```

```python
# hue -> style
```

```python
temp_df = gap[gap["country"].isin(["India","China","Pakistan"])]
sns.relplot(x="year",y="lifeExp",data=temp_df,kind="line",hue="country")
```

```python
temp_df = gap[gap["country"].isin(["India","Brazil","Germany"])]
sns.relplot(x="year",y="lifeExp",data=temp_df,kind="line",hue="country",style="continent")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",row="sex")
```

```python
# col wrap
sns.relplot(x="lifeExp",y="gdpPercap",kind="scatter",data=gap)
```

```python
sns.relplot(x="lifeExp",y="gdpPercap",data=gap,col="year",col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# plotting univariate histogram
sns.histplot(x="total_bill",data=tips)
```

```python
sns.displot(x="total_bill",kind="hist",data=tips)
```

```python
# bins parameter
sns.displot(x="total_bill",kind="hist",data=tips,bins=10)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables
```

```python
tips.head()
```

```python
sns.histplot(x="day",data=tips)
```

```python
# hue parameter
sns.histplot(x="tip",data=tips,hue="sex")
```

```python
# element -> step
sns.histplot(x="tip",data=tips,hue="sex",element="step")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.histplot(x="age",data=titanic)
```

```python
sns.histplot(x="age",data=titanic,hue="sex",element="step")
```

```python
# faceting using col and row
sns.displot(x="total_bill",kind="hist",data=tips,col="sex")
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(x="total_bill",data=tips)
```

```python
sns.displot(x="total_bill",data=tips,kind="kde")
```

```python
sns.displot(x="total_bill",data=tips,kind="kde",height=10)
```

```python
# hue -> fill
```

```python
sns.displot(x="total_bill",data=tips,kind="kde",hue="sex")
```

```python
sns.displot(x="total_bill",data=tips,kind="kde",hue="sex",fill=True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
```

```python
sns.kdeplot(x="total_bill",data=tips)
```

```python
sns.kdeplot(x="total_bill",data=tips)
sns.rugplot(x="total_bill",data=tips)
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color
```

```python
sns.histplot(x="total_bill",y="tip",data=tips)
```

```python
sns.displot(x="total_bill",y="tip",data=tips,kind="hist")
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
```

```python
sns.kdeplot(x="total_bill",y="tip",data=tips)
```

```python
sns.displot(x="total_bill",y="tip",data=tips,kind="kde")
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
```

```python
gap.head()
```

```python
temp_df = gap.pivot(index="country",columns="year",values="lifeExp")
temp_df
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(temp_df)
```

```python
# annot
```

```python
plt.figure(figsize=(25,25))
temp_df = gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
sns.heatmap(temp_df,annot=True)
```

```python
# linewidth
plt.figure(figsize=(25,25))
temp_df = gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

```python
# cmap
plt.figure(figsize=(25,25))
temp_df = gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap="summer")
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.
```

```python
iris = px.data.iris()
```

```python
iris.head()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
