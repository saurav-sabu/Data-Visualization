# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10 * x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ["eng","science","maths","sst","hindi"]
plt.pie(data,labels=subjects)
```

```python
# dataset
df = pd.read_csv("gayle-175.csv")
df.head()
```

```python
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%.2f")
```

```python
# percentage and colors
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%.2f",colors=["red","blue","black","green","pink","orange"])
```

```python
# explode shadow
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%.2f",colors=["red","blue","black","green","pink","orange"],explode=[0.2,0,0,0,0,0])
```

```python
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%.2f",colors=["red","blue","black","green","pink","orange"],explode=[0.2,0,0,0,0,0],shadow=True)
```
