### Colored Scatterplots

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
```

```python
iris["Species"] = iris["Species"].map({
    "Iris-setosa":0,
    "Iris-versicolor":1,
    "Iris-virginica":2
})
```

```python
iris.sample(5)
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
```

```python
# cmap
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="winter")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
```

```python
# colorbar
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="winter")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
# alpha
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="winter",alpha=0.2)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
y
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z  = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(10,10))

ax = plt.subplot(projection="3d")

ax.plot_surface(xx,yy,z,cmap="summer")
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(10,10))

ax = plt.subplot(projection="3d")

ax.plot_surface(xx,yy,z,cmap="summer")
```

```python
z = np.tan(xx) + np.cos(yy)

fig = plt.figure(figsize=(10,10))

ax = plt.subplot(projection="3d")

ax.plot_surface(xx,yy,z,cmap="summer")
```
